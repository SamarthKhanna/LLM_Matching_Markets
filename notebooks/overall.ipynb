```python
import pandas as pd
import csv
```

```python
parts = {
    '1': {
        'name': 'Achieving',
        'metrics': ['Correctness', 'Instability', 'Optimality'],
    },
    '2b': {
        'name': 'Resolving',
        'metrics': ['Correctness', 'Instability', 'Optimality'],
    },
    '2': {
        'name': 'Evaluating',
        'metrics': ['Correctness'],
    },
    '3': {
        'name': 'Preference',
        'metrics': ['Correctness'],
    },
}

combined = [['Model', 'Task', 'Metric', 'Value']]
file_names = {
    'Gemini-2.0-F': 'gemini20',
    'Llama-3.3-70B': 'llama33',
    'Qwen-QwQ-32B': 'qwen_qwq',
    'DeepSeek-70B': 'deepseek_dist',
    'OA-o3-mini': 'o3-mini',
    'DeepSeek-R1': 'deepseek',
    'Gemini-2.5-P': 'gemini25',

}


for model in file_names:
    for part in parts:
        suffix = 'cleaned' if part  == '3' else 'corrected'
        model_data = pd.read_csv(f'../evaluating_responses/part_{part}/{file_names[model]}_{suffix}.csv')
        metrics_data = None
        if part not in ['2','3']:
            model_data = model_data[model_data['Remarks'] == 'Processed smoothly.']
            metrics_data = pd.read_csv(f'../evaluating_responses/part_{part}/combined_instability_intersection.csv')
        for metric in parts[part]['metrics']:
            score = 0
            if metric == 'Correctness':
                for size in [10, 20, 50]:
                    model_data_size = model_data[model_data['Size'] == size]
                    score += sum(model_data_size[metric].values)*100/len(model_data_size)
                score = round(score/3, 2)
            else:
                model_metrics = metrics_data[metrics_data['Model'] == model][metric]
                score += sum(model_metrics.values)/len(model_metrics)
            print(model, part, metric, score)
            combined.append([model, parts[part]['name'], metric, score])


with open('combined_results.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(combined)
```
